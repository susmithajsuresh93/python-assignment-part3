```python
# Part B — Read
# ─────────────────────────────────────────

# step 1 - read file and print each line numbered
file  = open("python_notes.txt", "r", encoding="utf-8")
lines = file.readlines()
file.close()

print()
for i in range(len(lines)):
    line = lines[i].strip()   # removes \n from end
    print(str(i + 1) + ". " + line)

print()


# step 2 - total number of lines
print("Total lines:", len(lines))
print()


# step 3 - ask user for keyword and search
keyword = input("Enter a keyword to search: ")

found = False
for line in lines:
    # lower() makes the search case insensitive
    if keyword.lower() in line.lower():
        print(line.strip())
        found = True

if found == False:
    print("No lines found with the keyword: " + keyword)
```

```python
# Task 2 — API Integration
import requests

# Step 1 — Fetch and Display 20 Products

response = requests.get("https://dummyjson.com/products?limit=20")
data     = response.json()
products = data["products"]

print("ID  | Title | Category | Price | Rating")
print("--------------------------------------------")

for p in products:
    print(str(p["id"]) + " | " + p["title"] + " | " + p["category"] + " | $" + str(p["price"]) + " | " + str(p["rating"]))

print()
```

```python
# Step 2 — Filter Rating >= 4.5, Sort by Price

# filter first
filtered = []
for p in products:
    if p["rating"] >= 4.5:
        filtered.append(p)

# sort using sorted() with key
filtered = sorted(filtered, key=lambda p: p["price"], reverse=True)

print("Rating >= 4.5, sorted by price:")
for p in filtered:
    print(p["title"] + " | $" + str(p["price"]) + " | " + str(p["rating"]))

print()
```

```python
# Step 3 — Search by Category (laptops)

response = requests.get("https://dummyjson.com/products/category/laptops")
laptops  = response.json()["products"]

print("Laptops:")
for laptop in laptops:
    print(laptop["title"] + " - $" + str(laptop["price"]))

print()
```

```python
# Task 3 — Exception Handling
# Part A — Guarded Calculator

def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

# testing all three cases
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
print()
```

```python
# Part B — Guarded File Reader

def read_file_safe(filename):
    try:
        file    = open(filename, "r", encoding="utf-8")
        content = file.read()
        file.close()
        return content

    except FileNotFoundError:
        print("Error: File '" + filename + "' not found.")

    finally:
        # this runs every time - whether file was found or not
        print("File operation attempt complete.")


# Case 1: file exists
print(read_file_safe("python_notes.txt"))

# Case 2: file does not exist
print(read_file_safe("ghost_file.txt"))
```

```python
# Part C — Robust API Calls
# wrapping all requests calls in try-except

def get_products():
    url = "https://dummyjson.com/products"

    try:
        response = requests.get(url, timeout=5)
        data     = response.json()
        print("Products fetched successfully!")
        return data

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Error:", e)


def post_product(product_data):
    url = "https://dummyjson.com/products/add"

    try:
        response = requests.post(url, json=product_data, timeout=5)
        result   = response.json()
        print("Product added! ID assigned:", result["id"])
        return result

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")

    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")

    except Exception as e:
        print("Error:", e)
```

```python
# Part D — Input Validation Loop
import requests

while True:

    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ")

    # stop the loop if user types quit
    if user_input == "quit":
        print("Goodbye!")
        break

    # check if input is a valid number
    if not user_input.isnumeric():
        print("Warning: Please enter a valid number!")
        continue

    # check if number is in range 1-100
    product_id = int(user_input)
    if product_id < 1 or product_id > 100:
        print("Warning: Number must be between 1 and 100!")
        continue

    # only make the api call if input passed all checks
    try:
        response = requests.get("https://dummyjson.com/products/" + str(product_id), timeout=5)

        # check status code - 404 means not found
        if response.status_code == 404:
            print("Product not found.")

        elif response.status_code == 200:
            product = response.json()
            print("Title :", product["title"])
            print("Price : $" + str(product["price"]))

        print()

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print("Something went wrong:", e)
```

```python
# Task 4 — Logging to File
import requests
from datetime import datetime

# ─────────────────────────────────────────
# helper function to write a log entry
# ─────────────────────────────────────────

def write_log(location, error_type, message):
    # get current timestamp
    time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # build the log line
    log_line = "[" + time + "] ERROR in " + location + ": " + error_type + " — " + message + "\n"

    # open in append mode so old logs are not deleted
    file = open("error_log.txt", "a", encoding="utf-8")
    file.write(log_line)
    file.close()

    print("Logged:", log_line)


# ─────────────────────────────────────────
# trigger 1 — ConnectionError
# using a fake url that does not exist
# ─────────────────────────────────────────

print("Trying unreachable URL...")

try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)

except requests.exceptions.ConnectionError:
    write_log("fetch_products", "ConnectionError", "No connection could be made")

except requests.exceptions.Timeout:
    write_log("fetch_products", "Timeout", "Request timed out")

except Exception as e:
    write_log("fetch_products", "UnknownError", str(e))

print()


# ─────────────────────────────────────────
# trigger 2 — HTTP 404
# product id 999 does not exist
# ─────────────────────────────────────────

print("Looking up product ID 999...")

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    # 404 is not a python exception - we check status_code manually
    if response.status_code == 404:
        write_log("lookup_product", "HTTPError", "404 Not Found for product ID 999")

    elif response.status_code == 200:
        product = response.json()
        print("Title :", product["title"])
        print("Price : $" + str(product["price"]))

except requests.exceptions.ConnectionError:
    write_log("lookup_product", "ConnectionError", "No connection could be made")

except requests.exceptions.Timeout:
    write_log("lookup_product", "Timeout", "Request timed out")

except Exception as e:
    write_log("lookup_product", "UnknownError", str(e))

print()


# ─────────────────────────────────────────
# read and print full contents of error_log.txt
# ─────────────────────────────────────────

print("Contents of error_log.txt:")
print("─" * 50)

file  = open("error_log.txt", "r", encoding="utf-8")
lines = file.readlines()
file.close()

for line in lines:
    print(line.strip())
```
